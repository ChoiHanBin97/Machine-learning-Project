# 코스피 거래 창

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def Stock_Realtime_Search():
    url = 'https://finance.naver.com/sise/lastsearch2.nhn'
    request = requests.get(url)
    soup = BeautifulSoup(request.text, 'html.parser')
    text = soup.select('table.type_5 > tr')
    a = []
    for i in range(len(text)):
        a.append(text[i].text.strip().replace('\t',''))
    a = list(filter(None, a))
    real = []
    for i in range(1, len(a)):
        real.append(list(filter(None,a[i].split('\n'))))
    data = pd.DataFrame(real, columns=a[0].split('\n'))

    issue = []
    for i in range(len(data['등락률'])):
        if abs(float(data['등락률'][i].replace('%', ""))) > 10:
            query = data['종목명'][i]
            url = 'https://search.naver.com/search.naver?&where=news&query=%s'%(query) + '&sm=tab_tmr&frm=mr&nso=so:r,p:all,a:all&sort=0'
            request = requests.get(url)
            soup = BeautifulSoup(request.text, 'html.parser')
            titles = soup.select('dt ._sp_each_title')
            titles[0]['href']
            titles[0].text
            news = []
            news = [
                {'title' : title.text,
                'url' : title['href']    
                }for title in titles 
                ]
            issue.append(news)
        else: issue.append('')    
    data['Issue'] = pd.Series(issue, index=data.index)
    data.to_excel('C:\\Users\\user\\Desktop\\Stock\\주식 실시간 검색어\\%s.xlsx'  % (str(time.strftime('%Y-%m-%d %H시%M분%S초'))), index = False, encoding='euc-kr')
```

```python
Stock_Realtime_Search()
```

### 한시간마다 자동으로 크롤링하기

```python
while True:
    Stock_Realtime_Search()
    time.sleep(3600)
```
